# Box Plot

```python
sns.boxplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

From the above box plot,
Others category: the ss_p of male: 
Initial value=49, Mean=67, percentile: Q1=61, Q2=67, Q3=77, Q4=89.
Others category: the ss_p of female:
Initial value=45, Mean=74, percentile: Q1=64, Q2=74, Q3=79, Q4=87

For the Others category, the ssc_p scores of male candidate starts at 49. Many male candidates have scored between the range (61 to 77). The Highest score among the male candidates is 89. The ssc_p scores of female candidate starts at 45, with one outlier of 41 mark . Many female candidates have scored between the range (64 to 79). The Highest score among the female candidates is 87. If we compare the overall performance of male and female candidates from ssc_b "Others", we can see the female candidate have scored more marks than male candidates in higher numbers. But the lowest mark is set by female candidate and highest mark is set by male candidate.

Central category: the ss_p of male: 
Initial value=41, Mean=65, percentile: Q1=60, Q2=65, Q3=72, Q4=87.
Central category: the ss_p of female:
Initial value=43, Mean=66, percentile: Q1=61, Q2=78, Q3=75, Q4=88

For the Central category, the ssc_p scores of male candidate starts at 41. Many male candidates have scored between the range (60 to 72). The Highest score among the male candidates is 87. The ssc_p scores of female candidate starts at 43. Many female candidates have scored between the range (61 to 75). The Highest score among the female candidates is 88. If we compare the overall performance of male and female candidates from ssc_b "Central", we can see the female candidate have scored more marks than male candidates in higher numbers. But the lowest mark is set by male candidate and highest mark is set by female candidate.

# Point Plot

```python
sns.pointplot(x='ssc_p',y='ssc_b',data=dataset,hue='gender')
plt.show()
```

```python
others=dataset[dataset['ssc_b']=='Others']
central=dataset[dataset['ssc_b']=='Central']
f=others[others['gender']=='F']['ssc_p']
m=others[others['gender']=='M']['ssc_p']
F=central[central['gender']=='F']['ssc_p']
M=central[central['gender']=='M']['ssc_p']
print('Others Female mean: ',f.mean())
print('Central Female mean: ',F.mean())
print('Others Male mean:',m.mean())
print('Central Male mean: ',M.mean())
```

point plot of seaborn, plots the mean under different categories. The above plot is the visualization of ssc_b category and gender category with respect to the ssc_p. The mean of male and female candidates from central and others has been plot. To verify the plot, the mean is calculated and verified with point plot. 

# Factor Plot (Catplot)

```python
sns.catplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender',kind='violin')
```

From the above catplot, visualization is done on the male and female candidates from central and others board with respect to ssc_p score. Catplot is the combinations of box plot and violin plot. The box plot shows the initial values, and quartile ranges, and the violin plot shows the distribution density of the data. 

Box plot indicates the initial values, and quartile ranges Q1, Q2, Q3 and Q4.

For the male candidates in others, the density increases from the range 58 to 68, again a slight increase in density appears at the range 78 to 82. For the female candidates in others, the density increases at the range 69 to 82 representing large data distribution over that range.

For the male candidates in central ssc_b, the data distribution density increases at the range 58 to 68. For the female candidates in central ssc_b, the data distribution density increases at the range 68 to 82. This increase in density indicates more students have scored marks around thar range.
